## 프로젝트를 들어가며

```python
어린시절부터 전 세계적으로 가장 큰 인기를 누리는 잉글랜드 프리미어 리그의 뜨거운 열정과 클럽 간의 치열한 경쟁에 매료되어 온 열렬한 팬으로서,

저는 단순한 관람을 넘어 리그의 상위권클럽들이 어떻게 치열한 경쟁 속에서 꾸준히 정상에 오를 수 있었는지를 지켜보며 궁금증을 품어왔습니다.

이 프로젝트는, 이러한 의문을 해답을 찾고자 프리미어리그 내에서 강력한 입지를 구축한 상위권 클럽들의 핵심 요인들을 도출하고 기존 강팀들의 성공전략을 벤치마킹하여,

이를 신생 클럽에 최적화된 마케팅 전략을 수립하는데 주력할 것입니다. 본 프로젝트를 통해 도출된 전략은 신생 클럽이 단기적으로는 리그에 성공적으로 안착하고,

장기적으로는 리그의 중심으로 도약하는 데 실질적인 도움이 되는 전략이 될 것입니다.
```

### 사용된 개발환경, 패키지, 툴, 응용프로그램

```python
프로그래밍 언어: Python3
데이터 수집: Selenium, BeautifulSoup,
데이터 전처리: pandas, numpy, matplotlib, seaborn
데이터 분석: pandas, numpy, matplotlib, seaborn
딥러닝 프레임워크: 텐서플로 Sequencial
한글처리 형태소 패키지: konlpy-Okt
머신러닝 패키지: sklearn
머신러닝 알고리즘: 로지스틱 회귀
LLM: llama3
```

##### 순위 최적값

###### 순위 최적값 함수1

```python
feature_columns = X_train.columns
def find_optimal_values(model, encoder, feature_columns, target_rank):

    # 탐색 범위 설정
    height_range = np.arange(170, 200, 1)
    profit_range = np.arange(50, 180, 10)
    age_range = np.arange(20, 40, 1)

    # 최적 값 탐색
    best_values = None
    min_diff = float('inf') # 최소 차이 (초기값은 무한대)

    for height in height_range:
        for profit in profit_range:
            for age in age_range:
                # 팀 이름은 임의로 설정 (예: "Tottenham Hotspur")
                team_encoded = encoder.transform(['Tottenham Hotspur'])[0]
                input_data = pd.DataFrame([[profit, age, height, team_encoded]], columns=feature_columns)
                predicted_rank = model.predict(input_data)[0]

                # 목표 순위와의 차이 계산
                diff = abs(predicted_rank - target_rank)

                # 최소 차이 갱신 및 최적 값 저장
                if diff < min_diff:
                    min_diff = diff
                    best_values = {'Height': height, 'Profit': profit, 'Age': age}

    return best_values
```
